```python
import pandas as pd
```

```python
with open("FINAL_LLAMA_PRIMED.out", "r") as f:
    la_pt1 = f.readlines()
    
with open("FINAL_LLAMA_UNPRIMED.out", "r") as f:
    la_upt1 = f.readlines()
```

```python
with open("results_llama_PRIMED_FINAL.txt", "r") as f:
    rl_pt1 = f.readlines()
    
with open("results_llama_UNPRIMED_FINAL.txt", "r") as f:
    rl_upt1 = f.readlines()
```

```python
llama_probs_results = []
for i in la_pt1[3:]:
    results = i.split(" ")
    prob_i = results[0]
    llama_answer = results[1][:-1]
    llama_probs_results.append([float(prob_i), llama_answer])
```

```python
llama_prompt_primed = []
for r in rl_pt1:
    prompt_prev = r.split(" ")[:-1]
    prompt = " ".join(prompt_prev)
    llama_prompt_primed.append(prompt)
```

```python
df_llama_probs_results = pd.DataFrame(llama_probs_results, columns = ["probs_primed", "result_primed"])
df_prompts_primed_prev = pd.DataFrame(llama_prompt_primed, columns = ["prompt"])
```

```python
df_prompt_primed = pd.concat([df_prompts_primed_prev, df_llama_probs_results.reindex(df_prompts_primed_prev.index)], axis=1)
```

```python
df_prompt_primed
```

```python
# For unprimed:

llama_probs_results_u = []
for i in la_upt1[3:]:
    results = i.split(" ")
    prob_i = results[0]
    llama_answer = results[1][:-1]
    llama_probs_results_u.append([float(prob_i), llama_answer])
    
llama_prompt_unprimed = []
for r in rl_upt1:
    prompt_prev = r.split(" ")[:-1]
    prompt = " ".join(prompt_prev)
    llama_prompt_unprimed.append(prompt)
    
df_llama_probs_results_u = pd.DataFrame(llama_probs_results_u, columns = ["probs_unprimed", "result_unprimed"])
df_prompts_unprimed_prev = pd.DataFrame(llama_prompt_unprimed, columns = ["unprimed_prompt"])

df_prompt_unprimed = pd.concat([df_prompts_unprimed_prev, df_llama_probs_results_u.reindex(df_prompts_unprimed_prev.index)], axis=1)
```

```python
df_prompt_unprimed
```

```python
# Concatenating
```

```python
df_merged = pd.concat([df_prompt_primed, df_prompt_unprimed.reindex(df_prompt_primed.index)], axis=1)
```

```python
df_merged.head()
```

```python
df_master = pd.read_csv("mega_df.csv")
print(df_master.head())
```

```python
df_final = pd.merge(df_master, df_merged, on='prompt')
```

```python
df_final.head()
```

```python
df_merged_2 = df_final.copy()
```

```python
stable_mutable_relations = {"STABLE":["P30", "P495", "P740", "P103", "P364", "P19", "P20", "P279"],
                           "MUTABLE":["P108", "P39", "P937"]}
```

```python
stable_df = df_merged_2[df_merged_2['template'].isin(stable_mutable_relations['STABLE'])]

mutable_df = df_merged_2[df_merged_2['template'].isin(stable_mutable_relations['MUTABLE'])]

print("DataFrame of STABLE:")
print(stable_df.info())

print("\nDataFrame of MUTABLE:")
print(mutable_df.info())
```

```python
import numpy as np
```

```python
epsilon = float(1e-9)

stable_df['log_prob_ratio'] = np.log((stable_df['probs_primed']+epsilon) / (stable_df['probs_unprimed'] + epsilon))

print(stable_df)
```

```python
logprobratio_stable_avg = stable_df['log_prob_ratio'].mean()
```

```python
print(logprobratio_stable_avg)
```

```python
epsilon = 1e-9

mutable_df['log_prob_ratio'] = np.log((mutable_df['probs_primed'] + epsilon) / (mutable_df['probs_unprimed'] + epsilon))

print(mutable_df)
```

```python
logprobratio_mutable_avg = mutable_df['log_prob_ratio'].mean()
print(logprobratio_mutable_avg)
```
